```python

#You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using 
#the TreeNode class itself. Convert this binary tree into a binary search tree.
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def binary_tree_to_bst(root):
    # Step 1: Convert the binary tree to a sorted list
    sorted_list = inorder_traversal(root)

    # Step 2: Build a new binary search tree from the sorted list
    return build_bst(sorted_list)

def inorder_traversal(node):
    if node is None:
        return []
    return inorder_traversal(node.left) + [node.value] + inorder_traversal(node.right)

def build_bst(sorted_list):
    if not sorted_list:
        return None

    # Find the middle index
    mid = len(sorted_list) // 2

    # Create a new node with the middle value
    root = TreeNode(sorted_list[mid])

    # Recursively build the left and right subtrees
    root.left = build_bst(sorted_list[:mid])
    root.right = build_bst(sorted_list[mid+1:])

    return root
```

```python
#Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def find_distance(root, node1, node2):
    # Find the lowest common ancestor (LCA)
    lca = find_lca(root, node1, node2)

    # Calculate the distance from LCA to node1 and node2
    dist1 = find_distance_from_node(lca, node1)
    dist2 = find_distance_from_node(lca, node2)

    # Calculate the total distance
    total_distance = dist1 + dist2

    return total_distance

def find_lca(root, node1, node2):
    if root is None:
        return None

    # If both nodes are smaller than the root, move to the left subtree
    if node1 < root.value and node2 < root.value:
        return find_lca(root.left, node1, node2)

    # If both nodes are greater than the root, move to the right subtree
    if node1 > root.value and node2 > root.value:
        return find_lca(root.right, node1, node2)

    # If the nodes are on different sides of the root, the root is the LCA
    return root

def find_distance_from_node(node, target):
    if node is None:
        return 0

    # If the target value is less than the current node, move to the left subtree
    if target < node.value:
        return 1 + find_distance_from_node(node.left, target)

    # If the target value is greater than the current node, move to the right subtree
    if target > node.value:
        return 1 + find_distance_from_node(node.right, target)

    # If the target value is equal to the current node, return 0
    return 0
```

```python
#To convert a binary tree to a doubly linked list, we can use the Inorder traversal approach. Here's an example implementation in Python:

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

def convert_to_doubly_linked_list(root):
    if root is None:
        return None

    # Create a dummy node as the head of the doubly linked list
    dummy = DoublyLinkedListNode(None)

    # Initialize the previous node as None
    prev = None

    # Convert the binary tree to a doubly linked list using Inorder traversal
    inorder_traversal(root, dummy, prev)

    # Set the next of the last node to None
    curr = dummy.next
    curr.prev = None

    return curr

def inorder_traversal(node, dummy, prev):
    if node is None:
        return

    # Recursively convert the left subtree
    inorder_traversal(node.left, dummy, prev)

    # Create a new doubly linked list node for the current node
    curr = DoublyLinkedListNode(node.value)
    
    # Set the previous node of the current node
    curr.prev = prev
    
    if prev is None:
        # If prev is None, set the next of the dummy node to the current node
        dummy.next = curr
    else:
        # If prev is not None, set the next of the previous node to the current node
        prev.next = curr

    # Update the previous node to the current node
    prev = curr

    # Recursively convert the right subtree
    inorder_traversal(node.right, dummy, prev)
```

```python
#Write a program to connect nodes at the same level.
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.next = None

def connect_nodes_at_same_level(root):
    if root is None:
        return

    # Perform a level order traversal using BFS
    queue = [root]

    while queue:
        level_size = len(queue)

        # Connect nodes at the same level
        for i in range(level_size):
            node = queue.pop(0)

            # Connect the current node to the next node in the queue
            if i < level_size - 1:
                node.next = queue[0]

            # Enqueue the left and right children of the current node
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

# Example usage:
# Create a binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Connect nodes at the same level
connect_nodes_at_same_level(root)

# Print the connected nodes
current = root
while current:
    level_nodes = []
    while current:
        level_nodes.append(current.value)
        current = current.next
    print(level_nodes)
    current = current.left
```
